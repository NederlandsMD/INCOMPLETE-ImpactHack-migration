```python
# Dependencies
import requests
from config import api_key
import time
import pprint
```

```python
from newsapi import NewsApiClient

country = "Somalia"
issue = "drought"

url = (f'https://newsapi.org/v2/everything?'
       'q={country}%20{issue}&'
       'sortBy=popularity&'
       'apiKey={api_key}')
response = requests.get(url).json()

weNeed = response['articles']
pprint.pprint(weNeed)
```

```python
# Retrieve articles
articles_dict = {}
for i in range(0,4):
    articles_dict['article' + str(i+1)] = {"Title":weNeed[i]["title"], "Desc": weNeed[i]["description"], "Source": weNeed[i]["source"]["name"],
                      "Published": weNeed[i]["publishedAt"], "Link": weNeed[i]["url"]}
pprint.pprint(articles_dict)    
```
